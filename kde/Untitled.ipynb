```python
from sklearn.neighbors import KernelDensity
import numpy as np
X = np.array([[-1, -1], [-2, -1], [-3, -2], [1, 1], [2, 1], [3, 2]])
kde = KernelDensity(kernel='gaussian', bandwidth=0.2).fit(X)
kde.score_samples(X)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
lower, upper = -1, 6
data = (upper - lower)*np.random.rand(4) + lower
for i, d in enumerate(data):
    X = np.linspace(lower,upper,500)
    Y = norm.pdf(X, loc=d, scale=1)
    knl = plt.plot(X,Y,color='green', label='gaussian'+str(i+1))

# KDE
n = 4
h = 0.24
Y_pdf = []
for x in X:
    Y = 0
    for d in data:
        Y += norm.pdf((x-d)/h, loc=0, scale=1)
    Y /= n*h
    Y_pdf.append(Y)
kde = plt.plot(X,Y_pdf, color='red', label='KDE')
plt.scatter(data, [0]*len(data), color='red', marker='x')
plt.ylim([-0.05, 1])
plt.legend()
```
