```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ptick
import matplotlib.cm as cm
import numpy as np

## direct model
depth_1 = np.loadtxt("./6qubit_direct_1_cost.txt")
depth_2 = np.loadtxt("./6qubit_direct_2_cost.txt")
depth_3 = np.loadtxt("./6qubit_direct_3_cost.txt")
depth_4 = np.loadtxt("./6qubit_direct_4_cost.txt")
depth_5 = np.loadtxt("./6qubit_direct_5_cost.txt")
depth_6 = np.loadtxt("./6qubit_direct_6_cost.txt")

plt.figure(figsize=(12,8))
plt.rcParams["font.size"] = 14
plt.plot(depth_1, color=cm.hsv(0/10.0), label="depth=1")
plt.plot(depth_2, color=cm.hsv(1/10.0), label="depth=2")
plt.plot(depth_3, color=cm.hsv(2/10.0), label="depth=3")
plt.plot(depth_4, color=cm.hsv(3/10.0), label="depth=4")
plt.plot(depth_5, color=cm.hsv(4/10.0), label="depth=5")
plt.plot(depth_6, color=cm.hsv(5/10.0), label="depth=6")
plt.xlim(-5, 40)
plt.xlabel("Iteration")
plt.ylabel("Energy expectation value")
plt.title("direct method")
plt.legend()
plt.show()
```
